#### Question 1 - Read in the data set and print the % empty values for each column. Are there any columns with missing values?

```python
data_dir = 'groundhog.csv'
```

```python
ground_df = pd.read_csv(data_dir)
```

```python
new_df = ground_df.isnull().sum()
```

```python
col_names = ground_df.columns
print('Empty percent %:')
for i in range(len(col_names)):
    if i != 0:
        percent = (new_df[i]/ground_df.count()[i])*100
        print(col_names[i],percent)
```

#### Question 4 - Using the original dataframe (unchanged from Questoins 2, 3), create a mapping for the "Punxsutawney Phil" column which changes the categories into numerical quantities. For example: 'No Record' can be 0, 'No Shadow' can be 1, 'Partial Shadow' can be 2, and 'Full Shadow' can be 3. You should create the mapping *and* apply it to the data set by adding a new column called "Shadow".

```python
sha_rec = {'No Record':0,'No Shadow':1,'Partial Shadow':2,'Full Shadow':3}
ground_df['Shadow'] = ground_df['Punxsutawney Phil'].map(sha_rec)
ground_df = ground_df.drop(['Unnamed: 0'], axis=1)
```

```python
ground_df
```

#### Question 5 - For all rows which have no record, remove the rows which have empty values in all of the quantitative columns. (Only 9 should be removed.)
