## Problem 1

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as mpy
```

```python
students = pd.read_csv("student_class.csv")
students.head()
```

```python
stuMed = students.groupby(['Majors']).median()
majors = ['BNFO','CS','IS','IT','MATH']
x_index = np.arange(len(majors))
w = .25

mpy.bar(x_index-w,stuMed['Midterm'],width = w,label='Midterm',color='red')
mpy.bar(x_index,stuMed['Final'],width=w,label='Final',color='green')
mpy.bar(x_index+w,stuMed['Final Average'],width=w,label='Final Average',color='blue')
mpy.gcf().set_size_inches(15,7)
mpy.title('Median Midterm, Final, and Final Average Scores')
mpy.xlabel('Major')
mpy.ylabel('Median Score')
mpy.xticks(x_index,labels=majors)
mpy.legend()
mpy.show()
```

In this bar plot the data items represented are a student's scores of the Midterm, Final and Final Average
The attributes of this bar graph are the Midterm, Final, and Final Average
The marks of the bar graph are the three bars representing the grades
The channels are the colors of red, green, and blue as well as the horizontal and vertical position which represent the majors and score value, respectively.

When looking at the bar graph it can be said that BNFO is doing the best because they have the greatest score but the median can be shifted with ease in the data set so the data may be misleading. However, if the data is accurate then following the bar graph, BNFO and MATH majors seem to be doing the best in terms of score in comparison to other majors.

```python
students['Numeric Bin']=students['Numeric Bin'].astype('int')
bins=sorted(students['Numeric Bin'].unique())
bins.append(100)

mpy.hist(students['Final Average'],bins=bins,align='mid')
mpy.gcf().set_size_inches(15,7)
mpy.title('Number of Students in Each Grade Bin')
mpy.xlabel('Grade')
mpy.ylabel('Frequency')
mpy.show()
```

Identify the data items, attributes, and the marks and channels used. Critique the chart below. Is there
anything you would change? Is there anything you would improve? Also comment on the
distribution (approximately normal, skewed, etc)

The data items are a student's total overall grade.
The attriutes in the histogram is the final average.
The marks in the histogram are the bars representing the group of students in that grade score.
The channels used are the horizontal and vertical positions representing the grade group and the number of students in that group, respectively.

The axis of the chart are alright as is, however the title does not clearly represent what the data is trying to project.
```python
mpy.scatter(students['Midterm'],students['Final Average'],color='red',label='Midterm')
mpy.scatter(students['Final'],students['Final Average'],label='Final')
mpy.gcf().set_size_inches(14,7)
mpy.title('Exam vs Final Average')
mpy.xlabel('Exam Score')
mpy.ylabel('Final Average')
mpy.legend()
mpy.show()
```

The items in the scatter plot is a student.
The attributes in the scatter plot are the Final Average, Midterm, and Final
The marks used are the points.
The channels used in the scatter plot are the red and blue colors which represent the Midterm and Final scores, respectively as well as the horizontal and vertical positions which represent the exam score and Final Average score, respectively.

Us as the viewer of this visualization cannot clearly determine if a correlation exist between the data sets. A possible solution is to add a line of best fit and to seperate the data sets in order to show more accurate results.
